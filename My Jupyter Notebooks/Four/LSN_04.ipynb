We wish to simulate molecular dynamics on a relatively small (108 components) system

We make use of the Lennard Jones potential to simulate weakly-interacting noble gases.

We shall in particular simulate a system of Argon molecules ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu) in the following conditions:

1. solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
2. liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
3. gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

And we'll plot the progressive average & uncertainties for the total internal energy per particle E/N, the potential energy per particle U/N, the kinetic energy per particle K/N, the temperature T and pressure P (in natural units)

Let's observe what happens if we input the above mentioned values and start the simulation

```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize = (10,20))

mydata = open("Solid/total_energy.dat", "r")
length = len(mydata.readlines())
mydata.close()

#The following lines are my best attempt at converting .dat file entries into something I can work with

df = pd.read_csv('Solid/total_energy.dat', delimiter='\t')
datalist = df.values.tolist()

averages = []
error = []
j = []

for i in range (length-1):
    myrow = datalist[i]
    myrow = [x.strip(' ') for x in myrow]
    mylist = myrow[0].split()
    averages.append(float(mylist[2]))
    error.append(float(mylist[3]))
    j.append(i+1)

plt.subplot(5,1,1)

plt.errorbar(j, averages, error, xerr = None, fmt = ".r")    
plt.xlabel('block number')
plt.ylabel('Energy per particle (redux. u.)')
plt.grid(True)
plt.tight_layout()

mydata = open("Solid/potential_energy.dat", "r")
length = len(mydata.readlines())
mydata.close()

#The following lines are my best attempt at converting .dat file entries into something I can work with

df = pd.read_csv('Solid/potential_energy.dat', delimiter='\t')
datalist = df.values.tolist()

averages = []
error = []
j = []

for i in range (length-1):
    myrow = datalist[i]
    myrow = [x.strip(' ') for x in myrow]
    mylist = myrow[0].split()
    averages.append(float(mylist[2]))
    error.append(float(mylist[3]))
    j.append(i+1)
    
plt.subplot(5,1,2)

plt.errorbar(j, averages, error, xerr = None, fmt = ".g")    
plt.xlabel('block number')
plt.ylabel('Potential energy per particle (redux. u.)')
plt.grid(True)
plt.tight_layout()

mydata = open("Solid/kinetic_energy.dat", "r")
length = len(mydata.readlines())
mydata.close()

#The following lines are my best attempt at converting .dat file entries into something I can work with

df = pd.read_csv('Solid/kinetic_energy.dat', delimiter='\t')
datalist = df.values.tolist()

averages = []
error = []
j = []

for i in range (length-1):
    myrow = datalist[i]
    myrow = [x.strip(' ') for x in myrow]
    mylist = myrow[0].split()
    averages.append(float(mylist[2]))
    error.append(float(mylist[3]))
    j.append(i+1)

plt.subplot(5,1,3)

plt.errorbar(j, averages, error, xerr = None, fmt = ".r")    
plt.xlabel('block number')
plt.ylabel('Kinetic energy per particle (redux. u.)')
plt.grid(True)
plt.tight_layout()

mydata = open("Solid/temperature.dat", "r")
length = len(mydata.readlines())
mydata.close()

#The following lines are my best attempt at converting .dat file entries into something I can work with

df = pd.read_csv('Solid/temperature.dat', delimiter='\t')
datalist = df.values.tolist()

averages = []
error = []
j = []

for i in range (length-1):
    myrow = datalist[i]
    myrow = [x.strip(' ') for x in myrow]
    mylist = myrow[0].split()
    averages.append(float(mylist[2]))
    error.append(float(mylist[3]))
    j.append(i+1)

plt.subplot(5,1,4)
plt.errorbar(j, averages, error, xerr = None, fmt = ".m")    
plt.xlabel('block number')
plt.ylabel('Temperature (redux. u.)')
plt.grid(True)
plt.tight_layout()

mydata = open("Solid/pressure.dat", "r")
length = len(mydata.readlines())
mydata.close()

#The following lines are my best attempt at converting .dat file entries into something I can work with

df = pd.read_csv('Solid/pressure.dat', delimiter='\t')
datalist = df.values.tolist()

averages = []
error = []
j = []

for i in range (length-1):
    myrow = datalist[i]
    myrow = [x.strip(' ') for x in myrow]
    mylist = myrow[0].split()
    averages.append(float(mylist[2]))
    error.append(float(mylist[3]))
    j.append(i+1)

plt.subplot(5, 1, 5)
plt.errorbar(j, averages, error, xerr = None, fmt = ".c")    
plt.xlabel('block number')
plt.ylabel('Pressure (redux. u.)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

We can observe in all cases that the temperature will drift and follows the exact same curve shape as the kinetic energy. In the gas phase in particular it's reasonable that the molecules start moving away from each other and be less interacting due to increasing distances and gain kinetic energy compared to the initial crystalline state. Therefore the temperature rises for a transient before stabilizing

However the equilibrium temperatures are **not** what we want them to be therefore we need to explore various starting $T^*$ values to reach the desired equilibrium temperature

We find that to stabilize the temperature after 20000 steps (0.1 picoseconds) to the desired values initially mentioned an initial $T^* \approx 1.50 - 1.55$ is needed for the solid and $T^* \approx 1.9 - 2.0$ for the liquid phase

We then re-initialize the systems after reaching equilibrium wiith the desired temperatures to observe their statistical equilibrium properties

We observe initial fluctuations probably due to the imprecision of the restart process while using Verlet's algorithm but the quantities are nevertheless stabilized
