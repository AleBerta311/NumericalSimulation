#### Exercise 03.1 (European option pricing)

Black-Scholes theory provides an analitic expression for the "call" and "put" european options prices, modeled as exhibiting Geometric Brownian Motion (GBM)

Black-Scholes <span style="color:blue"> analytic solution </span> for the price at time t<T:
- European Call-option price:
$$ C\left[ S(t),t \right] = S(t) N(d_1) -K e^{-r(T-t)} N(d_2) $$
- European Put-option price:
$$ P\left[ S(t),t \right] = S(t) \left[ N(d_1)-1\right] -K e^{-r(T-t)} \left[ N(d_2)-1\right]$$

where 
$$d_1 = \frac{1}{\sigma\sqrt{T-t}}\left[ \ln \left( \frac{S(t)}{K} \right) + \left( r+\frac{\sigma^2}{2} (T-t) \right)\right]$$

where $d_2 = d_1 - \sigma\sqrt{T-t}$, and where
$$ N(x) = \int_{-\infty}^{x} dz \, e^{-\frac{z^2}{2}} = \frac{1}{2} \left[ 1+ {\rm erf} \left( \frac{x}{\sqrt{2}}\right) \right]$$
is the cumulative of Gaussian $\sim \mathcal{N}(\mu=0,\sigma^2=1)$,
and where
$${\rm erf}(x) = \frac{2}{\sqrt{\pi}}\int_{0}^{x} dz \, e^{-z^2}$$

We will use the following parameters: 
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$

We can calculate the expected price of the call and put options at time 0 (when the contract is made) according to the abovementioned equations:

```python
import scipy
from scipy.special import erf
import numpy
from numpy import exp
from numpy.lib.scimath import log, sqrt

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" == BLACK-SCHOLES ==")
    print("call: ", call)
    print("put: ", put)
```

We can obtain an estimation of the options' prices as $ \textlangle e^{-rt} (S(T)-K)^+ \textrangle$ averaging on 5000 values in each block and over 200 blocks overall

- To evaluate S(t) we can proceed in steps or we can directly evaluate the value at time T
- First we will perform a direct evaluation of $S(T)$

For the call options:

```python
import matplotlib.pyplot as plt
```

```python
mydata = open("call_prices.txt")

length = len(mydata.readlines())

mydata.close()

mydata = open("call_prices.txt")

call_prices = []
call_stds = []
j = []


for i in range (length):
    datastring = mydata.readline()
    datalist = datastring.split()
    call_prices.append(float(datalist[2]))
    call_stds.append(float(datalist[1]))
    j.append(i)

mydata.close()


plt.figure(figsize =(11, 8))
plt.errorbar(j, call_prices, call_stds, xerr = None, fmt = '.b')
plt.axhline(call, 0, 1000, color = "red", linestyle = "-")
plt.xlabel('Block Number')
plt.ylabel('Cumul. average call price')
plt.grid(True)
plt.savefig('calls.png', dpi = 800)
plt.show()

```

- The red line indicates the value estimated above using the analytical formula

```python
mydata = open("put_prices.txt")

length = len(mydata.readlines())

mydata.close()

mydata = open("put_prices.txt")

put_prices = []
put_stds = []


for i in range (length):
    datastring = mydata.readline()
    datalist = datastring.split()
    put_prices.append(float(datalist[2]))
    put_stds.append(float(datalist[1]))

mydata.close()

plt.figure(figsize =(11, 8))
plt.errorbar(j, put_prices, put_stds, xerr = None, fmt = '.c')
plt.axhline(put, 0, 1000, color = "red", linestyle = "-")
plt.xlabel('Block Number')
plt.ylabel('Cumul. average put price')
plt.grid(True)
plt.savefig('puts.png', dpi = 800)
plt.show()
```
