#### Ex. 2.1: Compute the following integral: $ \int_0^1 \frac{\pi}{2}cos(\frac{\pi x}{2})dx = 1 $

First we compute the integral by sampling a uniform distribution in [0, 1]

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.optimize import curve_fit
```

```python
mydata = open("results.txt", "r")
length = len(mydata.readlines())

mydata.close()

# the readline() method returns the line as a string, since each line contains 4 numbers we use the split() method
# to separate each line into a list of 2 elements and then we convert them to float to plot them.

mydata = open("results.txt", "r")

integrals = []
errors = []
j = []

for i in range (length):
    datastring = mydata.readline()
    datalist = datastring.split()
    integrals.append(float(datalist[0]))
    errors.append(float(datalist[1]))
    j.append(i)
    
    
plt.errorbar(j, integrals, errors, xerr = None, fmt = ".c")    
plt.xlabel('block number')
plt.ylabel('<integral>')
plt.grid(True)
plt.show()

mydata.close()    
```

Next we compute the same integral using importance sampling:

Let $ f(x) = \frac{\pi}{2} cos(\frac{\pi x}{2}) $

If we pretend not to know what the integral of this function over $[0,1] $ is, we could try using as our sampling $p(x) $ a function that looks similar to $f$ but that is certainly positive.

Let's pick $p(x) = |cos(\frac{\pi x}{2})| $ . 

Before we integrate this to normalize it, we can observe that the function inside the absolute value is always positive on $[0 ,1]$ so we can take  $p(x) = cos(\frac{\pi x}{2}) $

Unsurprisingly, the integral of this on $ [0, 1] $ evaluates to $\frac{\pi}{2} $ so it turns out our original function was already normalized and thus could have been used to evalute the integral with an error = 0. But if we know what the integral of $p(x)$ we also already know what the original integral is, without any need for numerical simulation.

Let's instead try taking another even simpler, less similar yet still positive and decreasing, function as our $p(x)$

Let $p(x) = e^{-x}$

Its integral over $[0, 1]$ is immediate to evaluate equal to $ 1 - e^{-1} = C  \approx 0.63212$ so $p(x) = \frac{1}{C}e^{-x}$

Thus we will evaluate $\int^1_0 \frac{f(x)}{p(x)} p(x) dx$ by sampling the exponential

```python
mydata = open("newresults.txt", "r")
length = len(mydata.readlines())

mydata.close()

# the readline() method returns the line as a string, since each line contains 4 numbers we use the split() method
# to separate each line into a list of 2 elements and then we convert them to float to plot them.

mydata = open("newresults.txt", "r")

integrals = []
errors = []
j = []

for i in range (length):
    datastring = mydata.readline()
    datalist = datastring.split()
    integrals.append(float(datalist[0]))
    errors.append(float(datalist[1]))
    j.append(i)
    
    
plt.errorbar(j, integrals, errors, xerr = None, fmt = ".r")    
plt.xlabel('block number')
plt.ylabel('<integral>')
plt.grid(True)
plt.show()

mydata.close() 
```

Notice how the errors have gotten significantly smaller despite not using the "best" sampling distribution

#### Exercise 02.2 Random 3D walk

First we shall simulate a discrete 3D walk on a lattice with spacing $ a = 1$.

At each discrete time the walker makes a forward or backward step of length equal to 1 in one of the 3 principal directions of the lattice: $x,y,z $

We will then plot the average value, over 100 blocks, of the distance from the origin after n steps, for every n in $[0, 100]$

- Note: to implement this discrete random walk two uniformly distributed numbers were generated for each step:
- The first number $n  \in [0,1)$ decides whether the walker moves forwards, if $ n \in [\frac{1}{2}, 1)$ or backwards otherwise
- The second number $n \in [0,1)$ decides whether the walker moves along the x axis, if $ n \in [0 , \frac{1}{3}) $ , y axis, if $n \in [\frac{1}{3}, \frac{2}{3})$ , or z axis otherwise  

```python
mydata = open("results_100_blocks_disc.txt")
length = len(mydata.readlines())
mydata.close()

mydata = open("results_100_blocks_disc.txt")

distances = []
errors = []
j = []

for i in range (length):
    datastring = mydata.readline()
    datalist = datastring.split()
    distances.append(float(datalist[0]))
    errors.append(float(datalist[1]))
    j.append(i)

plt.subplot(2,1,1)
plt.errorbar(j, distances, errors, xerr = None, fmt = '.-g')
plt.xlabel("Step number")
plt.ylabel("Distance from origin")
plt.title("Average distance from origin in 100 blocks after each step (disc)")
plt.grid(True)

def func(x, A, B):
    y = A*x**B
    return y
    
xdata = np.asarray(j)
ydata = np.asarray(distances)

parameters, covariance = curve_fit(func, xdata, ydata)

plt.subplot(2,1,2)
plt.title("sqrt(N)")
plt.plot(xdata, func(xdata, *parameters),'-r')

parameters

plt.show()
```

Now for the continuous variant:

At each discrete step the walker makes a movement to a point of sphere of radius 1, centered on its current position, chosen by random sampling of the azimuthal and polar angles $ \theta \in [0, \pi]$ and $\phi \in [0, 2\pi] $

We will then plot the average value, over 100 blocks, of the distance from the origin after n steps, for every n in $[0, 100]$

```python
mydata = open("results_100_blocks_cont.txt")
length = len(mydata.readlines())
mydata.close()

mydata = open("results_100_blocks_cont.txt")

distances = []
errors = []
j = []

for i in range (length):
    datastring = mydata.readline()
    datalist = datastring.split()
    distances.append(float(datalist[0]))
    errors.append(float(datalist[1]))
    j.append(i)


plt.subplot(2,1,1)
plt.errorbar(j, distances, errors, xerr = None, fmt = '.-c')
plt.xlabel("Step number")
plt.ylabel("Distance from origin")
plt.title("Average distance from origin in 100 blocks after each step (cont)")
plt.grid(True)


def func(x, A, B):
    y = A*x**B
    return y
    
xdata = np.asarray(j)
ydata = np.asarray(distances)

parameters, covariance = curve_fit(func, xdata, ydata)

plt.subplot(2,1,2)
plt.title("sqrt(N)")
plt.plot(xdata, func(xdata, *parameters),'-r')

plt.show()
parameters
```

There is great accordance between the two cases as in both we simulated a similar kind of 3D random walk

One can clearly observe that the graphs are fittable with a $ k \sqrt{N_{steps}}$ law with $k \approx 1$
