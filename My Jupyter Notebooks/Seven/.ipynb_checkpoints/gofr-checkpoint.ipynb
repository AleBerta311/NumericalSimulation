In the NSL simulator we have included tail corrections for potential energy, $U$, and pressure $P$.

### Exercise 07.2

We shall now use 500000 instantaneous U/N values with close to 50% acceptance ratio (0.06 delta) to estimate the autocorrelation of the energy estimation

```python
from statsmodels.graphics.tsaplots import plot_acf

import numpy as np
```

- solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
- liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
- gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

```python

fig, axs = plt.subplots(3, 1, figsize=(10, 12))
df = pd.read_csv('total_energy_solid.dat', delimiter='\t')
datalist = df.values.tolist()

length = len(datalist)
val = []
for i in range (length):
    myrow = datalist[i]
    myrow = [x.strip(' ') for x in myrow]
    mylist = myrow[0].split()
    val.append(float(mylist[1]))

values = np.array(val)


plot_acf(values, ax = axs[0], alpha = 0.05, title = 'Solid phase autocorrelation');

df = pd.read_csv('total_energy_liquid.dat', delimiter='\t')
datalist = df.values.tolist()

length = len(datalist)
val = []
for i in range (length):
    myrow = datalist[i]
    myrow = [x.strip(' ') for x in myrow]
    mylist = myrow[0].split()
    val.append(float(mylist[1]))

values = np.array(val)

plot_acf(values, ax = axs[1], alpha = 0.05, title = 'Liquid phase autocorrelation');

df = pd.read_csv('total_energy_gas.dat', delimiter='\t')
datalist = df.values.tolist()

length = len(datalist)
val = []
for i in range (length):
    myrow = datalist[i]
    myrow = [x.strip(' ') for x in myrow]
    mylist = myrow[0].split()
    val.append(float(mylist[1]))

values = np.array(val)

plot_acf(values, ax = axs[2], alpha = 0.05, title = 'Gas phase autocorrelation');
```

For the solid phase the autocorrelation is $\approx$ 0 after 60 steps the optimal block size would be around 600-800 steps

For the liquid phase the autocorrelation is $\approx$ 0 after 100 steps the optimal block size would be around 1000-1500 steps

For the gas phase the autocorrelation is $\approx$ 0 after 15 steps the optimal block size would be around 200-300 steps

Now we shall implement data blocking for blocks of different length and for each length print the value of the last rolling average uncertainty

```python
from math import sqrt

def error(L, filename):
    df = pd.read_csv(filename, delimiter='\t')
    datalist = df.values.tolist()
    length = len(datalist)
    N = int(length/L)   #number of blocks
    sum_prog = 0
    sum2_prog = 0
    err = 0
    sum = 0
    av = []
    av2 = []
    for i in range (N):
        sum = 0
        for j in range (i*L, (i+1)*L):
            myrow = datalist[j]
            myrow = [x.strip(' ') for x in myrow]
            mylist = myrow[0].split()
            sum+= float(mylist[1])
        av.append( sum/L )
        av2.append( (av[i])**2 )

    for i in range (N):
        sum_prog+= av[i]
        sum2_prog+= av2[i]
    sum_prog/= N 
    sum2_prog/= N

    err = sqrt(sum2_prog - sum_prog**2)/sqrt(N)

    return err
    
    
```

```python
filenames = ['total_energy_solid.dat', 'total_energy_liquid.dat', 'total_energy_gas.dat']

length = len(datalist)

sigma = []
j = []


for i in range (500):
    L = 10*(i+1)
    err = error(L, filenames[0])
    sigma.append(err)
    j.append(L)
```

```python
plt.figure(figsize=(20, 18))

plt.subplot(3,1,1)
plt.xlabel("Block length")
plt.ylabel("Error")
plt.title("$\sigma$, solid phase")
plt.plot(j, sigma, color = 'red', marker = '.', linestyle = 'dashed')
```

### Exercise 07.3

We show here an example of progressive $<g(r)>$ calculation of a solid state L-J system in the same conditions as initially mentioned

```python
fig, axs = plt.subplots(1, 1, figsize=(8, 5))
df = pd.read_csv('gofr.dat', delimiter='\t')
datalist = df.values.tolist()

length = len(datalist)
val = []
err = []
j = []
for i in range (length):
    myrow = datalist[i]
    myrow = [x.strip(' ') for x in myrow]
    mylist = myrow[0].split()
    val.append(float(mylist[2]))
    err.append(float(mylist[3]))
    j.append(i+1)

plt.title("Average global g(r)")
plt.xlabel("Block num")
plt.ylabel("<g(r)>")
plt.errorbar(j, val, err, xerr = None, fmt = '.r')
```
