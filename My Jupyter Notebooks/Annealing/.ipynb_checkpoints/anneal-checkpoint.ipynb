### Exercise 08.1:

First we write a program that can calculate $<H>$ of a trial $\Psi$ state for given $(\mu, \sigma)$ by sampling the $|\Psi|^2$ distribution using the Metropolis algorithm 

In order to make the algorithm start more efficiently we choose as starting point a position $x_0$ close to the minimum of the potential, where the peak of $\Psi$ will also be located

Given $V(x) = x^4 -\frac{5}{2}x^2$ and $\Psi(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}} + e^{-\frac{(x+\mu)^2}{2\sigma^2}} $ we chose values of $\mu = 1.1$ and $\sigma = 0.5$ as they seem like reasonable initial guesses

We execute 100000 sample moves in blocks of 5000 each and plot the progressive average and error

Then we use simulated annealing to find the optimal mu and sigma


```python
import matplotlib.pyplot as plt
import pandas as pd
```

### Exercise 08.2

We know use the simulated annealing (SA) algorithm to find the the closest-to-true values of $\mu$ and $\sigma$ (being those that minimize $<H>$ ) and therefore also of the energy

```python
plt.figure(figsize = (16,9))

mydata = open("Averages_per_temp.dat", "r")
length = len(mydata.readlines())
mydata.close()

df = pd.read_csv('Averages_per_temp.dat', delimiter='\t')
datalist = df.values.tolist()

beta = []
en = []


for i in range (length-1):
    myrow = datalist[i]
    myrow = [x.strip(' ') for x in myrow]
    mylist = myrow[0].split()
    en.append(float(mylist[1]))
    beta.append(1/float(mylist[0]))



plt.subplot(1,2,1)
plt.plot(beta, en, 'y')
plt.title("Average <H> at Temp")
plt.xlabel('$beta$')
plt.ylabel('Energy (red. units)')
plt.grid(True)
###########################################

mydata = open("<H>_estim.dat", "r")
length = len(mydata.readlines())
mydata.close()

df = pd.read_csv('<H>_estim.dat', delimiter='\t')
datalist = df.values.tolist()

j=[]
en = []


for i in range (length-1):
    myrow = datalist[i]
    myrow = [x.strip(' ') for x in myrow]
    mylist = myrow[0].split()
    en.append(float(mylist[1]))
    j.append(i+1)



plt.subplot(1,2,2)
plt.plot(j, en, 'y')
plt.title("<H> in each SA step")
plt.xlabel('$SA step$')
plt.ylabel('Energy (red. units)')
plt.grid(True)
#############################################
plt.tight_layout
plt.show()
```

We eventually converge to a solution at $\beta > 20$, that is after about 1200 SA steps

Note that to perform the Simulated Annealing we started with an initial guess of $\mu$ and $\sigma$ and then varied the temperature 40 times using a cooling rate of 0.91. That is, the temperature at step $k$ is $T_{k} = 0.91^{k}*T_0$ where $T_0 = 1.4$

Furthermore, the size of the SA step and the number of SA steps to make at each temperature also varied:

$ N_{steps} = T*100$

$\mu_{step} = 0.99^k*0.5$

$\sigma_{step} = 0.99^k*0.3$

This was done in order to allow for more exploration at higher "temperature" states and then drive towards convergence

The parameters of the solution found are $|\mu| = 0.87$ and $\sigma = 0.62 $ for $<H> = -0.430$

```python
plt.figure(figsize = (14,7))

mydata = open("metropolis_points.dat", "r")
length = len(mydata.readlines())
mydata.close()

df = pd.read_csv('metropolis_points.dat', delimiter='\t')
datalist = df.values.tolist()

points = []

for i in range (length-1):
    myrow = datalist[i]
    en.append(float(myrow[0]))
nbins = 250

plt.hist(en, nbins, density = True, label = 'sampled distribution')
plt.title("Sampled $|\Psi|^2$")
plt.xlabel('$block$')
plt.ylabel('density (red. units)')
#####################################################################
import numpy as np
def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

print("Ground state energy: ", E[0])

# Plot a few things
scale = 0.3
plt.plot(x, scale*V, color="Black", label="Potential") # plot the potential
plt.plot(x,(psi[0])**2, label = "Analytic G.S")
plt.xlabel("x")
plt.grid(True)
plt.xlim((-3,3))
plt.ylim((-0.6,0.6))

######################################################################
plt.legend(loc = 'upper right')
plt.grid(True)
plt.show()
```
