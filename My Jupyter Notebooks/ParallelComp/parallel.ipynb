We want to showcase the differences between a single GA search and 4 parallel searches with "Olympics", i.e. after each generation the processes evaluate the best of the best and receive it as their own

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize = (6,6))

mydata = open("Best_path_square3.dat", "r")
length = len(mydata.readlines())
mydata.close()

df = pd.read_csv('Best_path_square3.dat', delimiter='\t')
datalist = df.values.tolist()

x = []
y = []

for i in range (length-1):
    myrow = datalist[i]
    myrow = [x.strip(' ') for x in myrow]
    mylist = myrow[0].split()
    x.append(float(mylist[2]))
    y.append(float(mylist[3]))

plt.title("SINGLE GA (34 square), 74% steep, p 4% g 50% inv 10%, 250 Gen, Len = 10.49")
plt.plot(x,y, '.g')
plt.plot(x, y)
plt.plot(x[0], y[0], '.r')
```

What do we observe and learn from this? Essentially, that by running fewer generation but with parallel computation + exchange we can get a solution as good (Process 2 in this case) and one decent enough (Process 4) compared to one of the best solutions of the single process GA over almost twice as many generations.

Of course we could run it with many more processes and it would presumably perform even better thanks to the Olympics system but on my machine I was limited to 4 processes
